### 1) 간단한 그래프 그리기

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
```

pyplot 모듈은 그래프를 그리는 데 사용되는 다양한 함수를 포함하고 있음<br>
그중 plot 함수는 값을 서로 연결해서 라인 형태의 그래프를 그리는 함수

```python
x = range(0, 100)
y = [v*v for v in x]
plt.plot(x, y)
```

y = x^2 그래프

```python
x = range(0, 100)
y = [v*v for v in x]
plt.plot(x, y, 'ro')
```

인자 'ro' 에서 'r'은 red, 'o'는 그래프의 마커 모양을 의미

### Figure와 subplots
한 화면에 여러 개의 그래프를 그리려면 figure 함수를 통해 Figure 객체를 만든 후 add_subplot 메서드를 통해 그리려는 그래프 개수만큼 subplot을 만들면 됨<br>
add_subplot(2, 1, 1)<br>
add_subplot(2, 1, 2)<br>
(2, 1, 1) 2x1 행열의 subplot을 생성한다.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

x = range(0, 100)
y = [v*v for v in x]

ax1.plot(x, y)
ax2.bar(x, y)
```
